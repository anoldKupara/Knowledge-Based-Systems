```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bar_chart_race as bcr
from IPython.display import HTML
import warnings
warnings.filterwarnings("ignore")
```

```python
#Data preparation¶
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/country_vaccinations.csv')
data
```

```python
data.columns
```

```python
data["date"] = data["date"].apply(pd.to_datetime, dayfirst=True)
data2=data[['date','country','people_fully_vaccinated_per_hundred']].copy()
data2
```

```python
#Dataframe for Bar Chart Race¶
# set double indexes
data2.set_index(['date','country'])
```

```python
# unstack!!!
data3=data2.set_index(['date','country']).unstack()['people_fully_vaccinated_per_hundred']
data3
```

```python
data4 = data3.fillna(0)
data4
```

```python
#Create Bar Chart Race
df = data4
bcr.bar_chart_race(
    df=df,
    filename='Vaccination_BarChartRace.mp4',
    orientation='h',
    sort='desc',
    n_bars=10,
    fixed_order=False,
    fixed_max=False,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,'s': '',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func='median',
    period_length=1000,
    figsize=(5, 3),
    dpi=500,
    cmap='dark24',
    title='Fully Vaccinated per Hundred by Country',
    title_size=10,
    bar_label_size=7,
    tick_label_size=5,
    shared_fontdict={'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=True)
```
